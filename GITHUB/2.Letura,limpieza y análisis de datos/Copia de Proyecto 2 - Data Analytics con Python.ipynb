DIA CON MAYOR VENTA DEL PERIODO ESTUDIADO

```python
ventas_por_dia = Data_Store.groupby(Data_Store['Order Date'].dt.date)['Sales'].sum()
```

```python
dia_mayor_venta = ventas_por_dia.idxmax()
valor_venta = ventas_por_dia.max()

#con idmax me muestra el dia al que pertenece el indice de la mayor venta y la siguiente linea es para ver la cantidad de la venta
```

```python
print('El dia con mayor venta fue el :', dia_mayor_venta, 'con:', '$',valor_venta )
```

TOP 10 DIAS CON MAYORES VENTAS

```python
ventas_por_dia = ventas_por_dia.sort_values(ascending=False)
#sigo utilizando ventas por dia y ordeno con sort en forma descendente con ascending=False
```

```python
Top10_Dias = ventas_por_dia.head(10)
Top10_Dias = pd.DataFrame({'Order Date': Top10_Dias.index, 'Sales': Top10_Dias.values})
#creo un dataframe con los valores index de cada fecha y values tomando de sales
```

```python
Top10_Dias.head(10)
```

```python

sns.set(style='darkgrid', palette='muted') #utilizo seaborn para darle un estilo diferente
#creo el grafico de puntos utilizando matplotlib
plt.figure(figsize=(15, 5))
plt.scatter(Top10_Dias['Order Date'], Top10_Dias['Sales'])
plt.xlabel('Order date')
plt.ylabel('Sales')
plt.title('Top 10 Días con Mayores Ventas')
plt.xticks(rotation=45, fontsize=10)
#con esta linea le coloco las cantidades arriba de cada punto
for order_date, sales in zip(Top10_Dias['Order Date'], Top10_Dias['Sales']):
    plt.text(order_date, sales, str(sales), ha='center', va='bottom', fontsize=10)
#con esto aseguro una visualizacion agradable
plt.tight_layout()
plt.show()
```

Ventas por semana

```python
ventas_semana = Data_Store.groupby(pd.Grouper(key='Order Date', freq='W'))['Sales'].sum() #ase utiliza pd.Grouper, para agrupar por fechas, ejemplo, semanas,dias etc

ventas_semana = pd.DataFrame({'Order Date': ventas_semana.index, 'Sales': ventas_semana.values}) #creando un dataframe con el constructor de pandas
ventas_semana.head()
```

```python
# Configurar el estilo y la paleta de colores
sns.set(style='darkgrid', palette='muted')

# Crear una figura
plt.figure(figsize=(15, 5))

# Crear el gráfico de puntos
plt.scatter(ventas_semana['Order Date'], ventas_semana['Sales'], color='skyblue', label='Ventas', marker='o')

#creando linea de tendencia
z = np.polyfit(range(len(ventas_semana)), ventas_semana['Sales'], 1) #funcion con la cual creo un ajuste de un polinomio grado 1
p = np.poly1d(z) #creo la funcion polinomica con nuestro ajuste z
plt.plot(ventas_semana['Order Date'], p(range(len(ventas_semana))), "g--", label='Línea de Tendencia')# teniendo  nuestra funcion polinomica P, puedo operar dando valores de x
#Conectar puntos
plt.plot(ventas_semana['Order Date'], ventas_semana['Sales'], linestyle='-', color='gray', alpha=0.5)

# Etiquetas de ejes y título
plt.xlabel('Fecha de la Semana')
plt.ylabel('Ventas')
plt.title('Ventas por Semana con Línea de Tendencia')

plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

Como lo muestra la grafica anterior gracias a nuestra linea de tendencia, podemos apreciar un crecimiento de nuestras ventas a lo largo del periodo estudiado, se puede predecir que para el futuro la ventas subiran si se mantienen buenas estretegias de marketing y ventas

Top10 productos con mas ventas
